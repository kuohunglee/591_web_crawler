```python
#匯入相關套件

from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from time import sleep 
import json
import os
import hashlib
import re
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests as rq
import re
```

```python
#讀取存好的房屋買賣連結json檔案

f=open("D:\\資策會期末專題資料\\房地產\\591\\台北市中古屋爬蟲\\591_house_Taoyuan_link_25885.json",'r',encoding='utf-8')
strjson = f.read()
f.close()
listL= json.loads(strjson)
```

```python
#匯入requests套件
import re 

#用session來提高爬蟲速度
session = rq.Session()

#建立假header來破解防爬蟲機制
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36'}

#建立總會整空list來放所有爬取到的資料
list_sum=[]

#計次用變數
num=-1

#以迴圈爬取各個房屋鏈結的頁面資訊
for u in listL[0:1000]:
#     sleep(1)
    response = session.get(u, headers=headers)
    soup = BeautifulSoup(response.text, "lxml")

    
    try:
    #房屋總價
        list_big =[]  #存放section1的詳細資料
        list_big2 =[] #存放section2的詳細資料
        pricer = soup.find('span',class_='info-price-num').text
        price = re.search(r"[0-9]+", pricer)[0]
    #房屋單價
        unit_pricer = soup.find('div',class_='info-price-per').text
        unit_price = re.search(r"[0-9]+\.?[0-9]+", unit_pricer)[0]
        price_dict = {"房價":price, "單價":unit_price} #建立價錢字典
    # print(price_dict)
    
    
#section1
        list_floor= [] #存放floor-key的詳細資料
        list_main= []  #存放floor-value的詳細資料
        for f in soup.find_all('div',class_='info-floor-key'):
             list_floor.append(f.text)
    # print(list_floor)

        for m in soup.find_all('div',class_='info-floor-value'):
             list_main.append(m.text)
    # print(list_main)

        for n in range(len(list_main)):
            x = [list_main[n],list_floor[n]] 
            list_big.append(x) #索引讀取並結合成[[],[]]的list以利後面轉成字典能夠變成key&value型態

#     sleep(1)
#section2
        list_addr=[]  #存放addr-value的詳細資料
        list_addrt=[] #存放addr-key的詳細資料
        for ad in soup.find_all('span',class_='info-addr-value'):
             list_addr.append(ad.text)
    # print(list_addr)

        for adt in soup.find_all('span',class_='info-addr-key'):
             list_addrt.append(adt.text)
    # print(list_addrt)
        for n in range(len(list_addrt)):
            x = [list_addrt[n],list_addr[n]] 
            list_big2.append(x) #索引讀取並結合成[[],[]]的list以利後面轉成字典能夠變成key&value型態




#section3
        list_housed = []  #存放house-value的詳細資料
        list_housedt = [] #存放detail-house-key的詳細資料
        list_big3 =[]     #存放section3的詳細資料
        for a in soup.find_all('div',class_='detail-house-value'):
            list_housed.append(a.text)
#         print(f"key長度{len(list_housed)}")

        count=0 #計算housedt數量用
        for b in soup.find_all('div',class_='detail-house-key'):
            list_housedt.append(b.text)
            count+=1
#         print(f"Title長度{count}")

        for c in range(count): #藉由housedt數量計算來確保是正確數量及對應的housed配對
            x = [list_housedt[c],list_housed[c]]
            list_big3.append(x) #索引讀取並結合成[[],[]]的list以利後面轉成字典能夠變成key&value型態
#         print(list_big3)
        dict_big=dict(list_big+list_big2+list_big3) #把三個sections的資料和再一起並轉換成dictionary
        price_dict.update(dict_big)                 #把dict_big資料加入倒price_dict後頭
        list_sum.append(price_dict)                 #把price_dict(完整單一筆房屋買賣頁面資訊)資料加入list_sum(用來建立DataFrame)
#         print(list_sum)
        num+=1
        print(f"計次{num}")
    except:
        continue
```

```python
#爬取後的list_sum轉成DataFrame

df_from_dict = pd.DataFrame(list_sum)
```

```python
#把DataFrame存成csv檔到指定路徑

df_from_dict.to_csv("D:\\資策會期末專題資料\\房地產\\591\\台北市中古屋爬蟲\\591_高雄中古屋房價_2020803\\591_Taoyuan_house_0_953.csv",index=False, encoding="utf-8")
```
